```python
import numpy as np
import pandas as pd
```

```python
def preprocess(df):

  # To remove unwanted columns from the dataset.
  while True:
    x= input("Enter column name to remove, else enter NA: ")
    if x!='NA':
      if x in df.columns:
        df.drop(x, axis=1, inplace= True)
      else:
        print('Invalid Column name')
        continue
    else:
      break

  # To change the data type of the continuous columns from object to int/float
  while True:
    a= input("Enter column name to change datatype, else enter NA: ")
    if a!='NA':
      if a in df.columns:
        df[a]=pd.to_numeric(df[a], errors='coerce')
        df.dropna(inplace=True)
        df.reset_index(drop=True, inplace=True)
      else:
        print('Invalid Column name')
        continue
    else:
      break

  # To update the datatype of the categorical columns to int.
  for i in df.columns:
    if df[i].dtype==object:
      unique_vals = df[i].unique()
      val_dict = {}
      for val in unique_vals:
        if not val in val_dict:
          val_dict[val] = len(val_dict)

    df[i].replace(val_dict, inplace=True)
```

```python
 #Function to train model and predict values.
def model_prep(df):
  from sklearn.model_selection import train_test_split
  from sklearn.model_selection import GridSearchCV
  import xgboost as xgb
  from sklearn.model_selection import GridSearchCV

  from sklearn.preprocessing import StandardScaler



  a= input('Enter target column: ')
  x=df.drop(a,axis=1)
  y=df[a]
  x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.2, random_state=1)

  scaler = StandardScaler()
  x_train_scaled = scaler.fit_transform(x_train)
  x_test_scaled = scaler.fit_transform(x_test)

  clf = xgb.XGBClassifier()
  parameters = {
      "eta"    : [0.05,0.1,0.15] ,
      "max_depth"        : [3,4,5],
      "min_child_weight" : [ 3,5,7],
      "gamma"            : [0.5,0.6,0.7],
      "colsample_bytree" : [0.5,0.6,0.7]
      }

  grid = GridSearchCV(clf,
                      parameters,
                      scoring='roc_auc',
                      cv=5)

  grid.fit(x_train_scaled, y_train)
  print('Classification score on train data: ',grid.best_score_)
  print('Classification score on test data: ',grid.score(x_test_scaled, y_test))
```

```python
def model(z):

  df= pd.read_csv(z)      #read the dataset and store it in the new variable
  print(df.sample(10))
  print(df.info())

  preprocess(df)
  model_prep(df)
```
