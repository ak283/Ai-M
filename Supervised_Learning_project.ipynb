```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from scipy.stats import zscore
```

1.4 a

```python
x=df4.drop(labels= 'Class',axis=1)
y=df4['Class']
```

```python
xscaled= x.apply(zscore)
```

1.4 b

```python
x_train, x_test, y_train, Y_test =train_test_split(xscaled,y, test_size =0.20, random_state=1 )
```

1.4 c

```python
nnn= KNeighborsClassifier(n_neighbors = 5, weights= 'distance')
```

```python
nnn.fit(x_train, y_train)
```

```python
predict=nnn.predict(x_test)
```

```python
predict
```

```python
nnn.score(x_test,Y_test)
```

1.4 d

```python
from sklearn import metrics
```

```python
cm=metrics.confusion_matrix(Y_test, predict)
```

```python
cm
```

```python
df_cm = pd.DataFrame(cm, index = [i for i in ["H","S","N"]],
                  columns = [i for i in ["Predict H","Predict S","Predict N"]])
```

```python
sns.heatmap(df_cm, annot=True)
```

1.5 a

```python
nnn2= KNeighborsClassifier(n_neighbors = 10, weights='uniform' )
nnn2.fit(x_train, y_train)
predict=nnn2.predict(x_test)
nnn2.score(x_test,Y_test)
```

```python
nnn3= KNeighborsClassifier(n_neighbors = 10, weights='distance' )
nnn3.fit(x_train, y_train)
predict=nnn3.predict(x_test)
nnn3.score(x_test,Y_test)
```

1.5 b
When we set n_neighbors = 10, weights='uniform', prediction score was only0.69
But we set weights='distance' for same n_neighbors the score improved and becomes 0.74

1.5 c
When n_neighbors was set to 5 and weights='distance', score was 0.71
but when we set n_neighbors to 10 for the same weights value, score was 0.74
which clearly indicates on changing n_neighbors value score got improved.
